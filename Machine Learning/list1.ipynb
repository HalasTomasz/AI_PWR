```python
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import numpy as np

# Generate some example data
X = np.random.rand(100, 10)  # 100 samples, 10 features

# Initialize t-SNE with desired parameters
tsne = TSNE(n_components=3, perplexity=30, random_state=42)

# Fit t-SNE to the data and transform it to 2D
X_embedded = tsne.fit_transform(df)

plt.figure(figsize=(8, 6))
plt.scatter(X_embedded[:, 0], X_embedded[:, 2])
plt.title('t-SNE Visualization')
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)
scaled_df = pd.DataFrame(scaled_data, columns=df.columns)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=8)
pca.fit(scaled_df)
transformed_data = pca.transform(scaled_df)
print(pca.explained_variance_ratio_)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.ensemble import RandomForestClassifier

X_train, X_test,Y_train, Y_test = train_test_split(transformed_data, target_binary, random_state=42, test_size=0.2)
```

```python
print("Training set shape (X):", X_train.shape)
print("Testing set shape (X):", X_test.shape)
print("Training set shape (y):", Y_train.shape)
print("Testing set shape (y):", Y_test.shape)
forest = RandomForestClassifier(n_estimators=100,random_state=42)
forest.fit(X_train,Y_train)
Y_pred = forest.predict(X_test)

print(classification_report(Y_test,Y_pred))
```

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

dbscan = DBSCAN(eps=0.3, min_samples=10)

clusters = dbscan.fit_predict(X_train)
plt.scatter(X[:, 0], X[:, 1], c=clusters, cmap='viridis')
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("DBSCAN Clustering")
plt.show()
```
